```python
import pandas as pd
import numpy as np
```

```python
assist09 = pd.read_csv("raw/assist09_skill_builder_data_corrected.csv", encoding = "ISO-8859-15")
```

```python
print("students: ")
print(assist09['user_id'].nunique())
```

```python
pd.set_option('display.max_columns', 500)
assist09.head()
```

```python
assist09.describe() # 379938 rows
```

```python
# no gap time, we can fill it with the same number
df = assist09[['user_id', 'skill_id', 'problem_id', 'correct', 'ms_first_response','hint_count', 'skill_name']]
df = df.dropna()
df.describe() # 325637 rows
```

```python
df.head(1)
```

```python
df['ms_first_response'] = df['ms_first_response'] / 1000 # ms -> s
df['ms_first_response'] = df['ms_first_response'].round().astype(int)
df['skill_id'] = df['skill_id'].astype(int)
df['hint_count'] = df['hint_count'].astype(int)
df.head(1)
```

```python
df['difficulty'] = df.groupby('problem_id')['correct'].transform('mean')
df['difficulty'] = df['difficulty'].round(2)
df.head()
```

```python
# 按照正确率降序排序，并选择前10个
top_10_problems = df.sort_values(by='difficulty', ascending=True).head(10)

# 打印10个正确率最高的 problem id
print("10个正确率最高的 problem id:")
print(top_10_problems[['problem_id', 'difficulty']])
```

```python
problem_id_1_rows = df[df['problem_id'] == 94092]

# 打印所有 'problem_id' 为94092的行
print("所有 'problem_id' 为94092的行:")
problem_id_1_rows
```

```python
df['gap_time'] = 1
df.head()
```

```python
df = df.groupby('user_id', as_index=False).apply(lambda x: x.reset_index(drop=True))
```

```python
df.head()
```

```python
# remove seq_len < 3
len = df.groupby('user_id').size()
valid_users = len[len >= 3].index
df = df[df['user_id'].isin(valid_users)]
df.describe() # 325181 rows
```

```python
df = df[df['ms_first_response'] > 0]
df.describe() # 324237 rows
```

```python
df.head()
```

```python
unique_skills = df['skill_id'].unique()
unique_problems = df['problem_id'].unique()
print(unique_skills.__len__())
print(unique_problems.__len__())
print(df['user_id'].nunique())
```

```python
skill_mapping = {skill_id: idx+1 for idx, skill_id in enumerate(unique_skills)}
problem_mapping = {problem_id: idx+1 for idx, problem_id in enumerate(unique_problems)}
```

```python
skill_mapping_with_names = {skill_id: (idx+1, df[df['skill_id'] == skill_id]['skill_name'].iloc[0])
                            for idx, skill_id in enumerate(unique_skills)}
with open('assist09_skill_mapping.csv', 'w') as f:
    f.write(f"original_skill_id,new_skill_id,skill_name\n")
    for skill_id, (mapped_id, skill_name) in skill_mapping_with_names.items():
        f.write(f"{skill_id},{mapped_id},'{skill_name}'\n")
```

```python
df['skill_id'] = df['skill_id'].map(skill_mapping)
df['problem_id'] = df['problem_id'].map(problem_mapping)
```

```python
df.to_csv('raw/assist09_mapped.csv', index=False)
df.drop('skill_name', axis=1, inplace=True)
df.drop('problem_id', axis=1, inplace=True)
df.head()
```

```python
new_data = {
    'user_id': [],
    'skill_id': [],
    'correct': [],
    'ms_first_response': [],
    'hint_count': [],
    'difficulty': [],
    'gap_time': [],
}

max_answer_length = 100

for user_id, group in df.groupby('user_id'):
    skill_ids = []
    corrects = []
    responses = []
    hints = []
    diffs = []
    gaps = []

    # 遍历每个用户的答题记录
    for _, row in group.iterrows():
        skill_ids.append(str(int(row['skill_id'])))
        corrects.append(str(int(row['correct'])))
        responses.append(str(int(row['ms_first_response'])))
        hints.append(str(int(row['hint_count'])))
        diffs.append(str(row['difficulty']))
        gaps.append(str(int(row['gap_time'])))

    # 将problem_id和answer进行切片，保证每组的长度不超过最大回答长度
    skill_slices = [skill_ids[i:i + max_answer_length] for i in range(0, skill_ids.__len__(), max_answer_length)]
    answer_slices = [corrects[i:i + max_answer_length] for i in range(0, corrects.__len__(), max_answer_length)]
    response_slices = [responses[i:i + max_answer_length] for i in range(0, responses.__len__(), max_answer_length)]
    hint_slices = [hints[i:i + max_answer_length] for i in range(0, hints.__len__(), max_answer_length)]
    difficulty_slices = [diffs[i:i + max_answer_length] for i in range(0, diffs.__len__(), max_answer_length)]
    gap_slices = [gaps[i:i + max_answer_length] for i in range(0, gaps.__len__(), max_answer_length)]

    for i in range(skill_slices.__len__()):
        new_data['user_id'].append(user_id)
        new_data['skill_id'].append(','.join(skill_slices[i]))
        new_data['correct'].append(','.join(answer_slices[i]))
        new_data['ms_first_response'].append(','.join(response_slices[i]))
        new_data['hint_count'].append(','.join(hint_slices[i]))
        new_data['difficulty'].append(','.join(difficulty_slices[i]))
        new_data['gap_time'].append(','.join(gap_slices[i]))

new_df = pd.DataFrame(new_data)
new_df.head()
```

```python
new_df.to_csv('processed/assist09_processed.csv', index=False)
```
